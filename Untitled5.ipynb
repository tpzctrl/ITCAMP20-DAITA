# modeling

## tree decition

```python
x = dt.copy()
x = x.drop('Expert Diagnose', axis=1)
x
```

```python
y = dt['Expert Diagnose']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn import tree # decision tree
from sklearn.neighbors import KNeighborsClassifier # k-nn
from sklearn.naive_bayes import GaussianNB # naive bayes
```

```python
decision_tree1 = tree.DecisionTreeClassifier()
decision_tree1.fit(x_train, y_train)
```

```python
decision_tree1.get_depth()
```

```python
ydt = decision_tree1.predict(x_test)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(30,20))
plot_tree(decision_tree1, filled=True, feature_names=x_train.columns)
plt.title("Decision tree trained on all the iris features", fontsize=25)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
print("Decision Tree Accuracy:", accuracy_score(y_test, ydt))
```

F1-Score คือค่าความสมดุลระหว่างความแม่นยำในการทำนายในกลุ่มเป้าหมาย (Precision) และความสามารถในการตรวจจับเหตุการณ์ที่เราสนใจ (Recall) โดยสามารถคำนวณได้จาก 2x(PrecisionxRecall)/ 2x(Precision + Recall)

```python
from sklearn import metrics
print(metrics.classification_report(y_test, ydt))
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, ydt)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, )
plt.xlabel('true label')
plt.ylabel('predicted label')
```

## k-nn

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0, stratify=y)
```

```python
error_rate=[]#list that will store the average error rate value of k
for i in range (2,84):  #Took the range of k from 1 to 30
    clf=KNeighborsClassifier(n_neighbors=i)
    clf.fit(x_train,y_train)
    predict_i=clf.predict(x_test)
    error_rate.append(np.mean(predict_i!=y_test))

plt.figure(figsize=(30,10))
plt.plot(range(2,84),error_rate,marker="o",markerfacecolor="green",
     linestyle="dashed",color="red",markersize=15)
plt.title("Error rate vs k value",fontsize=20)
plt.xlabel("k- values",fontsize=20)
plt.ylabel("error rate",fontsize=20)
plt.xticks(range(2,84))
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(x_train, y_train)
```

```python
y_pred_knn = knn.predict(x_test)
```

```python
print("K-NN Accuracy:",accuracy_score(y_test, y_pred_knn))
```

```python
pd.crosstab(y_test, y_pred_knn)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred_knn))
```

```python
print(metrics.classification_report(y_test, y_pred_knn))
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, y_pred_knn)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, )
plt.xlabel('true label')
plt.ylabel('predicted label')
```
